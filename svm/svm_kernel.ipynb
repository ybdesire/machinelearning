# SVM with diffirent kernles

```python
import numpy as np
data_x = np.array([[-1, -1], [-2, -1], [1, 1], [2, 1]])
data_y = np.array([1, 1, 2, 2])
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear')
clf.fit(data_x, data_y) 
clf.score(data_x, data_y) # the accuracy at train dataset
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='poly')
clf.fit(data_x, data_y) 
clf.score(data_x, data_y) # the accuracy at train dataset
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='sigmoid')
clf.fit(data_x, data_y) 
clf.score(data_x, data_y) # the accuracy at train dataset
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='rbf')
clf.fit(data_x, data_y) 
clf.score(data_x, data_y) # the accuracy at train dataset
```
